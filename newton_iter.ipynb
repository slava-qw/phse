# Часть I. Метод Ньютона нахождения корней уравнения.

Напишите алгоритм, который выполняет итерации Ньютона для заданной функции $f(x)$ с известной производной $f'(x)$. Ваша функция должна находить корни $f(x)$ с заданной точностью $\epsilon$.

```python
def newton_iteration(f, fder, x0, eps=1e-5, maxiter=100):
    """Нахождение корней $f(x) = 0$ через итерации Ньютона.
    
    Parameters
    ----------
    f : callable
        Функция, корни которой мы хотим найти.
    fder : callable
        Производная `f`.
    x0 : float
        Начальное приближение итераций Ньютона.
    eps : float
        Заданная точность. 
        Алгоритм прекращает работу когда расстояние между последовательными приближениями меньше `eps`.
        По умолчанию 1e-5.
    maxiter : int
        Максимальное число итераций (по умолчанию 100).
        Алгоритм прекращается, когда число итераций достигает `maxiter`.
        Этот параметр нужен лишь для предотвращения бесконечного зацикливания.
    
    Returns
    -------
    x : float
        Найденное приближение к корню.
    niter : int
        Количество итераций.
    """
    x_np1 = x0
    x_n = x0 + 2 * eps
    i = 0

    while abs(x_np1 - x_n) >= eps and i < maxiter:
        x_n = x_np1
        x_np1 = x_n - f(x_n) / fder(x_n)

        i += 1
    x, niter = float(x_np1), i
    return x, niter
```

```python
from numpy.testing import assert_allclose

xx, nit = newton_iteration(lambda x: x**2 - 1, lambda x: 2.*x, x0=4)
assert_allclose(xx, 1.0, atol=1e-5)
assert nit < 10

xx, nit = newton_iteration(lambda x: x**2 - 1, lambda x: 2.*x, x0=-4)
assert_allclose(xx, -1.0, atol=1e-5)
assert nit < 10


from math import log, exp

xx, nit = newton_iteration(lambda x: exp(x)-2, lambda x: exp(x), x0=4, eps=1e-10)
assert_allclose(xx, log(2), atol=1e-10)
```

```python
_, max_nit = newton_iteration(lambda x: x**2 - 1, lambda x: 2.*x, x0=4)
xx = []
nit = []
for i in range(0, max_nit + 1):
  v_xx, v_nit = newton_iteration(lambda x: x**2 - 1, lambda x: 2.*x, x0=4, maxiter=i)
  xx.append(v_xx)
  nit.append(v_nit)

plt.figure(figsize=(12, 9))
plt.plot(nit, np.array(xx) - 1, label=r'$\log(\varepsilon)(n)$')
plt.yscale('log')
plt.legend()
plt.xlabel("Number of iterations (n)")
plt.ylabel(r"$\log(\varepsilon)$")
plt.grid()
plt.show()

# Сходимость метода Ньютона зависит от того, насколько близко к корню выбрано начальное приближение.
# Условие, гарантирующее сходимость: f'(x0) * f''(x0) > 0, где x0 - начальное приближение.
# Метод касательных имеет второй порядок сходимости (т.е. сходимость квадратичная)
```

# Часть II. Модифицированный метод Ньютона для кратных корней.

Рассмотрим функцию, имеющую кратный корень (например, $f(x) = (x^2 - 1)^2$). Реализуйте модифицированную итерацию Ньютона, 

$$
x_{n+1} = x_{n} - m \frac{f(x_n)}{f'(x_n)}
$$

```python
def mod_newton(f, fder, x0, m, eps=1e-5, maxiter=100):
    """Нахождение корней $f(x) = 0$ через итерации Ньютона.
    
    Parameters
    ----------
    f : callable
        Функция, корни которой мы хотим найти.
    fder : callable
        Производная `f`.
    x0 : float
        Начальное приближение итераций Ньютона.
    eps : float
        Заданная точность. 
        Алгоритм прекращает работу когда расстояние между последовательными приближениями меньше `eps`.
        По умолчанию 1e-5.
    maxiter : int
        Максимальное число итераций (по умолчанию 100).
        Алгоритм прекращается, когда число итераций достигает `maxiter`.
        Этот параметр нужен лишь для преcдотвращения бесконечного зацикливания.
    
    Returns
    -------
    x : float
        Найденное приближение к корню.
    niter : int
        Количество итераций.
    """
    x_np1 = x0
    x_n = x0 + 2 * eps
    i = 0

    while abs(x_np1 - x_n) >= eps and i < maxiter:
        x_n = x_np1
        x_np1 = x_n - m * f(x_n) / fder(x_n)

        i += 1
    x, it = float(x_np1), i
    return x, it
```

```python
for m in [1, 2, 3, 4, 5, 6]:
    xx, nit = mod_newton(lambda x: (x**2 - 1)**4,
                         lambda x: 4*(x**2 - 1)**3 * 2 * x,
                         x0=2, m=m, maxiter=10000, eps=1e-9)
    assert_allclose(xx, 1.0, atol=1e-8)
    
```

# Part III. Newton's fractal.

Consider the equation
$$ x^3 = 1 $$

It has three solutions in the complex plane, $x_k = \exp(i\, 2\pi k/ 3)$, $k = 0, 1, 2$.

The Newton's iterations converge to one of these solutions, depending on the starting point in the complex plane (to converge to a complex-valued solution, the iteration needs a complex-valued starting point).

Plot the **{basins of attraction}** of these roots on the complex plane of $x$ (i.e., on the plane $\mathrm{Re}x$ -- $\mathrm{Im}x$). To this end, make a series of calculations, varying the initial conditions on a grid of points. Color the grid in three colors, according to the root, to which iterations converged.


```python
# my own solution
def complex_nwet(f, fder, z0, eps=1e-5, maxiter=100):
    """Нахождение корней $f(z) = 0$ через итерации Ньютона.
    
    Parameters
    ----------
    f : callable
        Функция, корни которой мы хотим найти.
    fder : callable
        Производная `f`.
    z0 : Complex
        Начальное приближение итераций Ньютона.
    eps : float
        Заданная точность. 
        Алгоритм прекращает работу когда расстояние между последовательными приближениями меньше `eps`.
        По умолчанию 1e-5.
    maxiter : int
        Максимальное число итераций (по умолчанию 100).
        Алгоритм прекращается, когда число итераций достигает `maxiter`.
        Этот параметр нужен лишь для преcдотвращения бесконечного зацикливания.
    
    Returns
    -------
    x : Complex
        Найденное приближение к корню.
    niter : int
        Количество итераций.
    """
    x_np1 = z0
    x_n = z0 + 2 * complex(eps, eps)
    i = 0

    while abs(x_np1 - x_n) >= eps and i < maxiter:
        x_n = x_np1
        x_np1 = x_n - f(x_n) / fder(x_n)

        i += 1
    return x_np1, i


re_z = np.linspace(-2, 2, 1000)
im_z = np.linspace(-2, 2, 1000)
re_zz, im_zz = np.meshgrid(re_z, im_z)
roots = np.array([complex(1), complex(-0.5, np.sqrt(3) * 0.5), complex(-0.5, -np.sqrt(3) * 0.5)])
colors = {
    0: [],
    1: [],
    2: []
}

f = lambda z: z ** 3 - 1
fder = lambda z: 3 * z ** 2

for m in range(len(re_z)):
    for i in range(len(im_z)):
        z0 = complex(re_zz[m][i], im_zz[m][i])
        zz, _ = complex_nwet(f, fder, z0)
        epsilon = np.abs(roots - zz)
        j = np.where(epsilon == np.min(epsilon))
        colors[j[0][0]].append(np.array([z0.real, z0.imag]))

plt.figure(figsize=(12, 9))
plt.title(r'Newtons Basin of Attraction for $f(z) = z^3 - 1$')
for i in [0, 1, 2]:
    dots = np.array(colors[i])
    plt.scatter(dots[:,0], dots[:,1], marker='s', s=4)

a = np.array([1, -0.5, -0.5])
b = np.array([0, 0.5 * np.sqrt(3), -0.5 * np.sqrt(3)])
plt.scatter(a , b)
plt.show()
```

# Часть IV. Метод Ньютона для системы нелинейных уравнений.

Напишите алгоритм, который выполняет итерации Ньютона для заданной системы нелинейных уравнений $\vec{f}(x) = 0$ с известной матрицей Якоби $J(x)$. Ваша функция должна находить корни вектор-функции $\vec{f}(x)$ с заданной точностью $\epsilon$.

```python
def newton_system(F, jcb, X0, eps=1e-5, maxiter=100):
    """Нахождение корней $f(x) = 0$ через итерации Ньютона.
    from https://algowiki-project.org/ru/Метод_Ньютона_для_систем_нелинейных_уравнений
    
    Parameters
    ----------
    F : callable
        Вектор-функция системы, которую мы хотим решить.
    jcb : callable
        Якобиан `f`.
    X0 : array-like of floats, shape (n,)
        Начальное приближение итераций Ньютона.
    eps : float
        Заданная точность. 
        Алгоритм прекращает работу когда расстояние между последовательными приближениями меньше `eps`.
        По умолчанию 1e-5.
    maxiter : int
        Максимальное число итераций (по умолчанию 100).
        Алгоритм прекращается, когда число итераций достигает `maxiter`.
        Этот параметр нужен лишь для предотвращения бесконечного зацикливания.
    
    Returns
    -------
    X : array-like of floats, shape (n,)
        Найденное приближение к корню.
    niter : int
        Количество итераций.
    """
    x_np1 = X0
    x_n = X0 + 2 * eps
    i = 0

    while np.linalg.norm(x_np1 - x_n) >= eps and i < maxiter:
        x_n = x_np1
        dx = solve(jcb(x_n), -F(x_n))
        x_np1 = x_n + dx

        i += 1
    X, niter = x_np1, i
    return X, niter
```

```python
from numpy.testing import assert_allclose

def func(X):
    return np.array([X[0]**2 - 2*X[1]**4 + 1,
                     X[0] - X[1]**3 + 1])

def jac(X):
    return np.array([[2.*X[0], -8.*X[1]**3],
                     [1., -3.*X[1]**2]])

xx, nit = newton_system(func, jac, np.array([3., 2.]))
assert_allclose(xx, np.array([4.02879487, 1.71325223]), atol=1e-5)
assert nit < 10

xx, nit = newton_system(func, jac, np.asarray([1., 2.]))
assert_allclose(xx, np.asarray([-0.35119943,  0.86570596]), atol=1e-5)
assert nit < 10
```

Напишите функцию, возвращающую матрицу якобиана функции трех переменных `func2`:

```python
def func2(X):
    x, y, z = X
    return np.array([x ** 2 + y ** 2 + z ** 2 - 25,
                     x * y + y * z + z * x - 5,
                     x + y - 3])

def jac2(X):
    r"""Return the Jacobian of `func2(X)`.
    Suitable for all functions.
    
    
    The Jacobian matrix is defined as 
    
    $$
    J_{ij} = \partial f_i / \partial x_j
    $$
    
    so that the first row contains the derivatives of $f_0$
    with respect to the first, second etc coordinates; the second
    row contains the derivatives of $f_1$ with respect to
    the first, second etc coordinates; and so on.
    """
    x_v, y_v, z_v = X

    x, y, z = sp.symbols('x y z')
    f1, f2, f3 = func2([x, y, z])
    F = sp.Matrix([f1, f2, f3])
    
    J = F.jacobian([x, y ,z])
    # J_v = J.subs([(x, x_v), (y, y_v), (z, z_v)])
    
    s = (x, y, z)
    J_func = sp.lambdify(s, J, modules='numpy')
    J_ev = J_func(x_v, y_v, z_v)
    
    return J_ev
```
